```python
data = data.drop(columns=["lights"])
```

```python

X = data.drop(columns="Appliances")
Y = data["Appliances"]
```

```python
# Train-test split (70-30)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
scale = MinMaxScaler()
X_train_scaled = scale.fit_transform(X_train)
X_test_scaled = scale.transform(X_test)
```

```python
Reg_ = linear_model.LinearRegression()
Reg_.fit(X_train_scaled, y_train)
```

```python
train_predict = Reg_.predict(X_train_scaled)

from sklearn.metrics import mean_absolute_error
# Mean Absolute Error
MAE = mean_absolute_error(y_train, train_predict)
print(f"The Mean Absolute Error (Training Set) is: {round(MAE, 3)}")
```

```python
RMSE_train = np.sqrt(mean_squared_error(y_train, train_predict))
print(f"The Training RMSE is; {RMSE_train:.3f}")
```

```python
from sklearn.metrics import mean_absolute_error

# Predict on the test set
test_pred = Reg_.predict(X_test_scaled)

# Computing MAE
mae_test = mean_absolute_error(y_test, test_pred)

print(f"The Mean Absolute Error (test set) is; {round(mae_test, 3)}")
```

```python
rmse_test = np.sqrt(mean_squared_error(y_test, test_pred))
print(f"The Training RMSE is: {rmse_test:.3f}")
```

```python
#checking for possible overfitting
from sklearn.metrics import r2_score

train_r2 = r2_score(y_train, train_predict)
test_r2 = r2_score(y_test, test_pred)
```

```python
print("Train RÂ² Score:", round(train_r2, 3))
print("Test RÂ² Score:", round(test_r2, 3))
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()  # default alpha=1.0
ridge.fit(X_train, y_train)

y_pred_test = ridge.predict(X_test)

rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
print(f"The Test RMSE (Ridge) is:", rmse_test)
```

```python
from sklearn.linear_model import Lasso

# Assume X_train, y_train are already defined

lasso = Lasso()  # default alpha=1.0
lasso.fit(X_train, y_train)

# Get the coefficients (feature weights)
coefficients = lasso.coef_

# Count how many are non-zero
non_zero_count = sum(coeff != 0 for coeff in coefficients)

print("Number of features with non-zero weights:", non_zero_count)
```

```python
# Predict on test set
y_pred_test = lasso.predict(X_test)

# Calculate RMSE
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
print(f"The Test RMSE (Lasso) is:", rmse_test)
```
